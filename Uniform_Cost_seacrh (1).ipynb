```python
import heapq# # priority queue (min-heap) use karne ke liye library

# Graph representation (Adjacency List)
graph = {
    'S': [('A', 3), ('B', 2), ('C', 7)],
    'A': [('D', 8), ('E', 15)],
    'B': [('G', 20)],
    'C': [('G', 6)],
    'D': [],
    'E': [],
    'G': []
}

def uniform_cost_search(graph, start, goal):
    # Priority Queue (min-heap)
    pq = []
    heapq.heappush(pq, (0, start, [start]))  # (cost, node, path)

    visited = set()# visited nodes ko store karne ke liye
    # empty set (duplicate avoid karta hai)


    while pq:
        cost, node, path = heapq.heappop(pq)

        if node in visited:
            continue

        visited.add(node)

        # Goal check
        if node == goal:
            return cost, path

        # Explore neighbors
        for neighbor, weight in graph[node]:
            if neighbor not in visited:
                heapq.heappush(pq, (cost + weight, neighbor, path + [neighbor]))

    return None, None


# Run UCS
start = 'S'
goal = 'G'

cost, path = uniform_cost_search(graph, start, goal)# UCS function call ho
#raha hai jo shortest cost aur path return karta hai

print("Shortest Path:", " -> ".join(path))
print("Total Cost:", cost)
```

```python
import heapq  # priority queue (min cost nikalne ke liye use hoti hai)

# Graph define kar rahe hain (node : [(neighbor, cost)])
graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('D', 1), ('E', 3)],
    'C': [('F', 5)],
    'D': [('G', 2)],
    'E': [],
    'F': [],
    'G': []
}

def uniform_cost_search(graph, start, goal):
    pq = []
    # empty priority queue bana rahe hain jahan (cost, node, path) store hoga

    heapq.heappush(pq, (0, start, [start]))
    # start node ko 0 cost ke sath queue me daal diya

    visited = set()
    # visited set banaya taake repeat nodes avoid ho

    while pq:
        # jab tak queue empty nahi hoti loop chalega

        cost, node, path = heapq.heappop(pq)
        # sab se choti cost wala node nikal rahe hain

        if node in visited:
            continue
        # agar node already visit ho chuka hai to skip karo

        visited.add(node)
        # node ko visited me add kar rahe hain

        if node == goal:
            return cost, path
        # agar goal mil gaya to cost aur path return karo

        for neighbor, weight in graph[node]:
            # current node ke neighbors check kar rahe hain

            if neighbor not in visited:
                heapq.heappush(pq, (cost + weight, neighbor, path + [neighbor]))
                # neighbor ko updated cost ke sath queue me add kar rahe hain

    return None, None
    # agar goal na mile to None return hoga


# Start aur Goal define
start = 'A'
goal = 'G'

# function call
cost, path = uniform_cost_search(graph, start, goal)

# output print
print("Path:", " -> ".join(path))
print("Cost:", cost)
```
